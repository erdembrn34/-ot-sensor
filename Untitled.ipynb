```python
# üõ†Ô∏è 1Ô∏è‚É£ Gerekli K√ºt√ºphaneler
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import plotly.express as px
import os
from scipy import stats
```

```python
# üìå 2Ô∏è‚É£ Sahte IoT Veri √úretimi
num_sensors = 10
num_records = 500
start_time = datetime.now()

data = []
for i in range(num_records):
    for sensor_id in range(1, num_sensors+1):
        record = {
            "sensor_id": sensor_id,
            "temperature": round(np.random.normal(25, 5), 2),
            "humidity": round(np.random.normal(50, 10), 2),
            "pressure": round(np.random.normal(1013, 5), 2),
            "co2": round(np.random.normal(400, 50), 2),
            "light": round(np.random.normal(300, 50), 2),
            "battery": round(np.random.uniform(20, 100), 2),
            "vibration": round(np.random.normal(0.5, 0.1), 2),
            "status": np.random.choice(["OK", "WARN", "FAIL"]),
            "timestamp": start_time + timedelta(seconds=i)
        }
        data.append(record)

df = pd.DataFrame(data)
```

```python
# üìå 3Ô∏è‚É£ Eksik Veri ve Temizleme
df.loc[df.sample(frac=0.02).index, 'humidity'] = np.nan
df['humidity'] = df['humidity'].fillna(df['humidity'].mean())
```

```python
# üìå 4Ô∏è‚É£ Anomali / Outlier Detection (Z-score)
df['temp_zscore'] = df.groupby('sensor_id')['temperature'].transform(lambda x: np.abs(stats.zscore(x)))
anomalies = df[df['temp_zscore'] > 3]
```

```python
# üìä 5Ô∏è‚É£ Batch Analizi ve Rolling Metrics
agg_df = df.groupby("sensor_id")[["temperature","humidity","pressure","co2","light","battery","vibration"]].agg(
    ["mean","min","max"]
).reset_index()
```

```python
df['temp_rolling'] = df.groupby('sensor_id')['temperature'].transform(lambda x: x.rolling(10,1).mean())
df['humidity_rolling'] = df.groupby('sensor_id')['humidity'].transform(lambda x: x.rolling(10,1).mean())
```

```python
# üìå 6Ô∏è‚É£ Veri Kalitesi Kontrolleri
dq_report = pd.DataFrame({
    "sensor_id": df['sensor_id'].unique(),
    "missing_humidity": df.groupby('sensor_id')['humidity'].apply(lambda x: x.isna().sum()),
    "outliers_temp": df.groupby('sensor_id')['temp_zscore'].apply(lambda x: (x>3).sum())
})
```

```python
# üìÇ 7Ô∏è‚É£ Veri G√∂l√º / Parquet Kaydetme
if not os.path.exists("data"):
    os.makedirs("data")

df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day

df.to_parquet("data/sensor_data.parquet", partition_cols=["sensor_id","year","month","day"], index=False)
anomalies.to_csv("data/anomalies.csv", index=False)
agg_df.to_html("data/summary_report.html")
dq_report.to_html("data/data_quality_report.html")

print("T√ºm veriler ve raporlar kaydedildi:")
print(" - Parquet: data/sensor_data.parquet")
print(" - Anomalies CSV: data/anomalies.csv")
print(" - Summary HTML: data/summary_report.html")
print(" - Data Quality HTML: data/data_quality_report.html")
```

```python
# üìà 8Ô∏è‚É£ G√∂rselle≈ütirme (Matplotlib)
plt.figure(figsize=(12,6))
for sensor_id in df['sensor_id'].unique():
    sensor_data = df[df['sensor_id'] == sensor_id]
    plt.plot(sensor_data['timestamp'], sensor_data['temperature'], label=f'Sensor {sensor_id}')
plt.xlabel("Zaman")
plt.ylabel("Sƒ±caklƒ±k (¬∞C)")
plt.title("IoT Sens√∂r Sƒ±caklƒ±k Zaman Serisi")
plt.legend()
plt.show()
```

```python
# üìä 9Ô∏è‚É£ Interaktif Plotly Grafik
fig = px.line(df, x="timestamp", y="temperature", color="sensor_id",
              hover_data=["humidity","co2","battery","vibration","status"],
              title="IoT Sens√∂r Sƒ±caklƒ±k (Interaktif)")
fig.show()
```
